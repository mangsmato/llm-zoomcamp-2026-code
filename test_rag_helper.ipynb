```python
import os
from dotenv import load_dotenv
load_dotenv()  # take environment variables from .env.

from ingest import load_faq_data, build_index
from rag_helper import RAGBase
from google import genai
```

```python
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
client = genai.Client(api_key=GOOGLE_API_KEY)
```

```python
# test the RAGBase class
documents = load_faq_data()
index = build_index(documents)
rag_assistant = RAGBase(index=index, llm_client=client)
```

```python
li_st_of_prompts = [
    "Can I still join the course after it started?",
    "How do I get a certificate?",
    "What is the format of the DataTalks courses?",
    "Which is the best AI model",
]
```

```python
for prompt in li_st_of_prompts:
    response = rag_assistant.rag(prompt)   
    print(f"\n=== Prompt: {prompt} ===")
    print(f"Response:\n {response}")
```

```python
# override the instructions 
custom_instructions = """
You're a course teaching assistant.
Answer the QUESTION based on the CONTEXT from the FAQ database.
Use only the facts from the CONTEXT when answering the QUESTION.
""".strip()
rag_assistant = RAGBase(index=index, 
                        llm_client=client, 
                        instructions=custom_instructions
                        )
```

```python
for prompt in li_st_of_prompts:
    response = rag_assistant.rag(prompt)   
    print(f"\n=== Prompt: {prompt} ===")
    print(f"Response:\n {response}")
```
