```python
from ingest import load_faq_data
import time
from sqlitesearch import TextSearchIndex
```

```python
# load the data using the function from ingest.py
documents = load_faq_data()
print(f"Loaded {len(documents)} documents")
```

```python
# Filter to just the LLM Zoomcamp documents:
docs_llm = [doc for doc in documents if doc["course"] == "llm-zoomcamp"]
print(f"LLM Zoomcamp: {len(docs_llm)} documents")
```

```python
# Now create a sqlitesearch index and add documents one by one with a small delay
#  (to simulate slow ingestion):
index = TextSearchIndex(
    text_fields=["question", "section", "answer"],
    keyword_fields=["course"],
    db_path="faq.db"
)

c=0
for doc in docs_llm:
    index.add(doc)
    print(f"Iteration {c+1}: Added: {doc['question'][:60]} ...")
    time.sleep(0.5)
    c+=1

index.close()
print("Done. Index saved to faq.db")
```
